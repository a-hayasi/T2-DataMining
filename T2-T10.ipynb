# 1. Introdução
Criptomoedas são moedas digitais descentralizadas
* moedas, por serem utilizadas para transações monetárias, 
* digitais, por não serem físicas e circularem em redes digitais
* descentralizadas, pois não precisam de mediadores como bancos ou governos (entidade central) para validar suas transações.

Em 1998 surgiram ideias para moedas digitais como o B-money e Bit Gold. Porém, a primeira moeda digital descentralizada foi criada em 2009 por Satoshi Nakamoto, chamada de BitCoin. A Bitcoin foi projetada para ser uma carteira digital, onde cada transação feita é guardado o valor, a data, o vendedor e o comprador.
Inicialmente possuía um apelo mais ideológico que prático, principalmente para libertários por ser uma moeda descentralizada. Começou a ser usada em massa para comércio ilegal em mercados negros online, porém eventualmente começou a ser utilizada para poupança.

Países com problemas de cambio ou crises economícas, incentivaram o uso do BitCoin para ser utilizado como forma de poupança, elevando exponencialmente o valor da moeda. Bitcoins comprados por 50 doláres em 2010 se valorizaram para 40 milhões de doláres em 2017, se tornando uma criptomoeda extremamente atraente para investimentos em poupança.




# 3. Conjunto de Dados


O conjunto de dados utilizados foi o fornecido pelo [Coin Market Cap](https://coinmarketcap.com/pt-br/) um site internacional que possui diversos dados relevantes sobre mais de 2095 criptomoedas desde sua criação e um dos mais utilizados por investidores em criptomoedas.

Cada amostra do conjunto de dados é uma observação sobre uma criptomoeda em uma data específica. Ou seja, cada amostra possui:
* - sigla (symbol): Sigla da criptomoeda amostrada
* - nome (name): Nome da criptomoeda amostrada
* - date (data): Data, no formato AAAA/MM/DD, em que a amostragem foi realizada
* - ranknow : Rank da criptomoeda na data observada
* - open (abertura): Valor (em dólares) de abertura da criptomoeda na data.
* - high (alta): Valor (em dólares) mais alto da criptomoeda na data.
* - low (baixa): Valor (em dólares) mais baixo da criptomoeda na data.
* - close (fechamento): Valor (em dólares) da criptomoeda na última transação do dia.
* - volume: Quantidade (em dólares) de criptomoeda que foi negociada sobre 24hrs.
* - market (Capitalização de Mercado): É uma forma de ranquear o tamanho de uma criptomoeda. A capitalização de mercado refere-se ao produto da quantidade de moedas existentes com o seu valor na data.
* - close_ratio (Taxa de fechamento): Taxa aproximada de fechamento
* - spread : Diferença, em dólares, do valor mais alto e mais baixo registrados nos dia.

# 4. Análise de Impacto e Regressão (Experimentos)

A partir de todas as análises e informações coletadas até agora, é possível inferir que a Bitcoin é de longe a criptomoeda não apenas mais importante, como também, impactante, sendo seu valor de mercado altamente influenciador nas outras criptomoedas.

Além disso, o desenvolvimento de um algoritmo capaz de predizer, por exemplo, o valor que uma criptomoeda vai fechar em determinado período de tempo (ou naquele mesmo dia) pode ser vantajoso para diversas relações de transação. Contudo, apesar de audacioso, o desenvolvimento de tal algoritmo é extremamente complexo pois, assim como foi o caso das bolhas, as variações do preço de uma criptomoeda estão sempre relacionadas com fatores externos, sejam eles políticos ou comerciais.

Não o bastante, a falta de dados que descreva precisamente o comportamento das criptomoedas impede bruscamente o desenvolvimento de um regressor eficiente pra predizer valores de abertura e fechamento, por exemplo. Afinal, até 2017, a valorização das criptomoedas nunca foi tão alta quanto em 2017.


Porém, isto não impede o desenvolvimento de um regressor orientado a influência de uma criptomoeda sobre a outra. Isto é, dado o valor de abertura, alta e baixa de um Bitcoin é possível prever o valor de fechamento da Ethereum ou Bitcoin Cash? 
Essa informação pode ser útil em diversas situações de transações, trocas de moedas e diversas outras estratégias financeiras, tal como, verificar se o comportamento de uma criptomoeda pode descrever o de outra.

### Estratégias

Portanto, foi definido o seguinte objetivo: É possível determinar o preço de fechamento de uma criptomoeda em relação as informações de outra?

Para isso, foi desenvolvido um regressor treinado com os dados de abertura, baixa e alta de uma criptomoeda e, testado nos dados de todas as outras criptomoedas. Por exemplo, no primeiro caso, treinou-se o regressor com dados da Bitcoin e testou seu poder de predição para a Ethereum, Bitcoin Cash, Ripple e Litocoin.

O regressor adotado foi k Vizinhos Mais Próximos (em sua forma para regressão) cujo número de vizinhos adotado foi igual a 5. Já as métricas de avaliação adotadas foram:
* Coeficiente de Determinação ($R^2$): Mede o quão bem o modelo é capaz de prever valores futuros. Caso seja 1, então o modelo possui uma capacidade de generalização perfeita. Além disso, apesar de raro, o Coeficiente de Determinação pode ser negativo.
* Erro Quadrático Médio (MSE): O Erro Quadrático Médio é simplemente o somatório do valor de todos os erros do modelo, elevado ao quadrado. Como o MSE toma os valores de erro quadrados, a métrica é extremamente sensível a valores discrepantes, posto que os valores muito discrepantes levam a grandes erros.
* Pontuação de variação explicada (EVS): Essa métrica mede o quão bem um modelo consegue predizer variações no conjunto de dados. O valor mais alto que a EVS pode alcamçar é 1 (nesse caso, se tem um regressor com uma grande capacidade de prever variações) e valores mais baixos expressam um pior desempenho. 

## 4.2. BTC x ETH, BCH, XRP e LTC

O primeiro caso de treinamento e teste foi verificar a capacidade do Bitcoin de descrever o comportamento das outras criptomoedas. Portanto, 5-NN (kNN com 5 vizinhos mais próximos) foi treinado com os dados de abertura, alta, baixa e fechamento da Bitcoin em uma dia $x$ e testado com os dados de fechamento das demais criptomoedas, também no dia $x$.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics   import mean_squared_error, explained_variance_score

np.random.seed(1)
```

```python
x_train = btc.iloc[:,5:8]
y_train = btc.iloc[:,8]

x_test = [eth.iloc[:,5:8], xrp.iloc[:,5:8], bch.iloc[:,5:8], ltc.iloc[:,5:8]]
y_test = [eth.iloc[:,8], xrp.iloc[:,8], bch.iloc[:,8], ltc.iloc[:,8]]

clf = None
clf = KNeighborsRegressor(n_neighbors=5, weights='distance')
clf.fit(x_train, y_train)
```

```python
results = []

for x, y in zip(x_test, y_test):
        y_pred = clf.predict(x)
        r2  = clf.score(x, y)
        mse = mean_squared_error(y, y_pred)
        evs = explained_variance_score(y, y_pred)
        aux = [r2, mse, evs, y_pred]
        results.append(aux)
```

```python
for i in results:
    print('R²: {:.4f}\tMSE: {:.4f}\t EVS: {:.4f}'.format(i[0], i[1], i[2]))
```

```python
plt.figure(figsize=(15,7))
plt.plot(eth_dates, y_test[0],     color='#00a8ff')
plt.plot(eth_dates, results[0][3], color='#e84118')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(xrp_dates, y_test[1],     color='#8c7ae6')
plt.plot(xrp_dates, results[1][3], color='#fbc531')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(bch_dates, y_test[2],     color='#4cd137')
plt.plot(bch_dates, results[2][3], color='#ff9f1a')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.plot(ltc_dates, y_test[3],     color='#18dcff')
plt.plot(ltc_dates, results[3][3], color='#c56cf0')
plt.xticks(rotation=90)
plt.show()
```
